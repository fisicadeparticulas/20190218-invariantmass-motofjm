```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
datos = pd.read_csv('Dimuon_DoubleMu.csv')
```

```python
df = pd.DataFrame(datos)
```

```python
print (df)
```

```python
df['MassInvariant'] = ( (df.E1+df.E2)**2 - ( (df.px1+df.px2)**2 + (df.py1+df.py2)**2 + (df.pz1+df.pz2)**2) ) **.5
```

```python
print(df)
```

```python
print(df)
```

```python
df.to_csv('Calculo de masa.csv')
```

```python
fig = plt.figure(figsize=(24,12))
plt.scatter(df.index.values, df['M'])

plt.title ('Dimuon Mass')
plt.xlabel ('Eventos')
plt.ylabel ('Masa [GeV] ')
plt.show()
```
